#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
html=req.get(url).content

soup=bs(html, 'html.parser')
````

````python
py = soup.find_all('a')

url_py = []

for i in py:
    url_py.append(i.get('href',''))    # con el get le digo que me de href y con '' aunque no exista

    
    
def check_url(url):                            # Funcion que me devuelve un booleano 
    url = url.lower()
    return 'http' in url and 'python' in url


url_py_filter = list(filter(check_url, url_py))    # Hago un filter el que le paso el resultado de la función

url_py_filter[0:5]
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

### REVISAR Y VOLVER A HACER

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
html=req.get(url).content

soup=bs(html, 'html.parser')
````

````python
terremoto = soup.find_all('tbody', {'id': 'tbody'})[0]

filas=terremoto.find_all('tr')

filas=[f.text.strip().split('\n') for f in filas]
````

````python
final =[]

for i in filas:
    
    for j in i:
        
        e = j.split(u'\xa0')
        final.append(e)
````

````python

final2=[]

for f in final:
    
    tmp=[]
    
    for palabra in f:
        
        if palabra!='':
            
            tmp.append(palabra)
    
    final2.append(tmp)
````

````python

df = pd.DataFrame(final2)

df.head()
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
html=req.get(url).content

soup=bs(html, 'html.parser')
````

````python
# Esto hay que descarse la librería  unicode y cambiar el formato.


div = soup.find_all('div', {'class': 'central-featured'})[0]

lang=div.find_all('div')

language = []

for i in lang:
    i = i.text.split()
    i = ' '.join(i[:-1]).rstrip('+').replace(' ', '-', 1).replace(' ', '')
    
    language.append(i)
    
language
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
html=req.get(url).content

soup=bs(html, 'html.parser')
````

````python
# Encuentro la primera tabla 
tabla=soup.find_all('table')[3]

#tabla=soup.find_all('table')
````

#### Para wikipedia parece ser siempre lo mismo

````python
# Encuentro columnas y la meto en una lista con comphrension

filas=tabla.find_all('tr')

filas=[f.text.strip().split('\n') for f in filas]

filas[:5]
````

````python
# Doble bucle en la lista de listas para quitar lo elementos vacios ''

final=[]

for f in filas:
    
    tmp=[]
    
    for palabra in f:
        
        if palabra!='':
            
            tmp.append(palabra)
    
    final.append(tmp)
    
final[:5]
````

````python
# Columnas el primer elemento de la lista de listas y las filas el resto

col_names=final[0]

data=final[1:]

df=pd.DataFrame(data, columns=col_names)

df.head(10)
````

#### IMDB's Top 250 data (movie name, Initial release, director name and stars) as a pandas dataframe

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
html=req.get(url).content

soup=bs(html, 'html.parser')
````

````python
tabla=soup.find_all('table')[0]
````

````python
filas=tabla.find_all('tr')

filas=[f.text.strip().split('\n') for f in filas]

filas[:5]
````

````python
final=[]

for f in filas:
    
    tmp=[]
    
    for palabra in f:
        
        if palabra not in ['', ' ']:
            
            tmp.append(palabra.strip())
    
    final.append(tmp)
    
final[:5]
````

````python
col_names=['Rank', 'Title', 'Year','IMDb Rating']

data=pd.DataFrame(final[1:])

data = data[[0, 1, 2, 3]]
````

````python
data.columns = col_names

data
````
